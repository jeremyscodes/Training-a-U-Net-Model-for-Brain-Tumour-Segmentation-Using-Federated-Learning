```python
main_dir = '..\\..\\My_Results\\Federated_U-Net\\'
dir2 = '2_Clients_Full_train\\raw_data'
dir4 = '4_Clients_Full_train\\raw_data'
dir_8 = '8_Clients_Full_train\\raw_data'

import os
import matplotlib.pyplot as plt
import numpy as np
central_loss_path = main_dir + dir2 + "\\P_global_model_loss_2_clients_FedAvg.npy"

import numpy as np
import matplotlib.pyplot as plt

def moving_average(data, window_size):
    """Compute the moving average of the data."""
    cumsum = np.cumsum(data)
    cumsum[window_size:] = cumsum[window_size:] - cumsum[:-window_size]
    return cumsum[window_size - 1:] / window_size
def moving_average2(data, window_size):
    """Compute the moving average of the data with central alignment."""
    half_window = window_size // 2
    result = []
    
    for i in range(len(data)):
        start_idx = max(0, i - half_window)
        end_idx = min(len(data), i + half_window + (window_size % 2))
        
        avg_value = np.mean(data[start_idx:end_idx])
        result.append(avg_value)
    
    return result


def plot_metric_from_npy(file_path, num_clients, metric_name, window_size=2):
    # Load data from the .npy file
    data = np.load(file_path, allow_pickle=True)

    # Extract rounds and dice coefficients
    rounds = [entry[0] for entry in data]
    if metric_name != 'Loss':
        metric = [100.0 * entry[1] for entry in data]
    else: metric = [entry[1] for entry in data]

    # Compute the moving average if window_size > 1
    if window_size > 1:
        metric = moving_average(metric, window_size)
        rounds = rounds[window_size-1:]  # Adjust rounds to match the length of the smoothed metric

    # Plot the data
    plt.figure()
    plt.plot(rounds, metric)
    plt.grid()
    plt.ylabel(metric_name+" (%)")
    plt.xlabel("Round")
    plt.title(f"Aggregated model's {metric_name} with {num_clients} clients (Window size: {window_size})")
    plt.show()

# Example usage:
# plot_dice_from_npy('path_to_file.npy', 10, 'Dice Coefficient', window_size=5)


# Example usage:
plot_metric_from_npy(central_loss_path, 2, "Loss", window_size=1)

plot_metric_from_npy(central_loss_path, 2, "Loss", window_size=5)
plot_metric_from_npy(central_loss_path, 2, "Loss", window_size=10)

central_loss_4_clients = main_dir + dir4 + "\\A_global_model_loss_4_clients_FedAvg.npy"
plot_metric_from_npy(central_loss_4_clients, 4, "Loss", window_size=1)
plot_metric_from_npy(central_loss_4_clients, 4, "Loss", window_size=10)
```
